1: Histogram for a miles per gallon

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'D:\data analytics\data analysis\cars.csv'
cars_df = pd.read_csv(file_path, encoding='ISO-8859-1')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(cars_df['mpg'], kde=True, color='blue', bins=10)
plt.title('Distribution of Miles per Gallon (mpg)')
plt.xlabel('Miles per Gallon')
plt.ylabel('Frequency')
plt.show()


```

2. Visualizing the distribution of 'hp' (Horsepower)

```python
plt.figure(figsize=(10, 6))
sns.histplot(cars_df['hp'], kde=True, color='green', bins=10)
plt.title('Distribution of Horsepower (hp)')
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.show()
```

3. Visualizing the relationship between 'wt' (Weight) and 'mpg' (Miles per gallon)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=cars_df['wt'], y=cars_df['mpg'], hue=cars_df['cyl'], palette='coolwarm', s=100)
plt.title('Weight vs. Miles per Gallon (mpg)')
plt.xlabel('Weight (in 1000 lbs)')
plt.ylabel('Miles per Gallon')
plt.legend(title="Cylinders")
plt.show()
```

4. Visualizing the relationship between 'hp' (Horsepower) and 'mpg' (Miles per gallon)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=cars_df['hp'], y=cars_df['mpg'], hue=cars_df['cyl'], palette='coolwarm', s=100)
plt.title('Horsepower vs. Miles per Gallon (mpg)')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon')
plt.legend(title="Cylinders")
plt.show()
```

5. Visualizing the distribution of 'disp' (Displacement)

```python
plt.figure(figsize=(10, 6))
sns.histplot(cars_df['disp'], kde=True, color='purple', bins=10)
plt.title('Distribution of Displacement (disp)')
plt.xlabel('Displacement (cu. in.)')
plt.ylabel('Frequency')
plt.show()
```

6. Visualizing the count of different numbers of cylinders ('cyl')

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=cars_df['cyl'], palette='Set2')
plt.title('Count of Different Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

 7. Visualizing the distribution of 'drat' (Rear axle ratio)

```python
plt.figure(figsize=(10, 6))
sns.histplot(cars_df['drat'], kde=True, color='orange', bins=10)
plt.title('Distribution of Rear Axle Ratio (drat)')
plt.xlabel('Rear Axle Ratio')
plt.ylabel('Frequency')
plt.show()
```

 8. Visualizing the relationship between 'qsec' (1/4 mile time) and 'mpg' (Miles per gallon)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=cars_df['qsec'], y=cars_df['mpg'], hue=cars_df['am'], palette='coolwarm', s=100)
plt.title('1/4 Mile Time vs. Miles per Gallon (mpg)')
plt.xlabel('1/4 Mile Time (seconds)')
plt.ylabel('Miles per Gallon')
plt.legend(title="Transmission (0 = Auto, 1 = Manual)")
plt.show()
```

9. Visualizing the count of different transmission types ('am')

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=cars_df['am'], palette='Set3')
plt.title('Count of Different Transmission Types (0 = Automatic, 1 = Manual)')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.show()
```

10. Visualizing the count of different gear types ('gear')

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=cars_df['gear'], palette='Set1')
plt.title('Count of Different Gear Types')
plt.xlabel('Number of Gears')
plt.ylabel('Count')
plt.show()
```

 11. Visualizing the count of different carburetor types ('carb')
